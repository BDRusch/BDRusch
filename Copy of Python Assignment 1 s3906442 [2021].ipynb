```
#defining cumsum
def cumsum(mylist):     #creating function for cumsum based on arbitrary list
  sumcum = [mylist[0]]    #sumcum is the list of cumulatative sums, starting with the first element in the list
  for i in range(len(mylist)):  #for loop thatll iterate as many times as there are items in the list
    if i == (len(mylist)-1):    #stopping code before list[i+1] exceeds len(list)
      break
    sumcum.append(sumcum[i] + mylist[i+1])  #(sumcum[i] + list[i+1]) makes the newest cumulatative sum
  return sumcum


#defining integers
integers = [i+1 for i in range(100)]   #i+1 to have integers start from 1 rather than 0 (could also do range(1,100,1))
print("integers: ", integers) 
print("cumsum: ", cumsum(integers))
print()

#defining alt_integers
alt_integers = [i for i in range(1,1000,2)] #odd integers starting from 1
print("alt_integers:", alt_integers)
print("cumsum: ", cumsum(alt_integers))
print()

#defining fibonacci
fibonacci = [1,1]
for i in range(100):
  fibonacci.append(fibonacci[i]+fibonacci[i+1])
print("fibonacci: ", fibonacci)
print("cumsum: ", cumsum(fibonacci))
```
