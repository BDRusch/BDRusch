```
#Excersize 2
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as rd

for n in range(4):#using a for loop to obtain the power of the number of samples 
  
  x = rd.uniform(low = -1, high = 1, size = 10**(n))#numpy.random.uniform is used to obtain random values of x between -1 and 1 with 10**n samples
  y = rd.uniform(low = -1, high = 1, size = 10**(n))#numpy.random.uniform is used to obtain random values of y between -1 and 1 with 10**n samples
  inside, outside = [[],[]] , [[],[]]. #obtain a list for the values inside the circle and those outside of it with two lists inside for the x and y values
  for i in range(len(x)):#use two for loops to iterate all the y values over each x value for the formula of a circle with radius equal to 1
    for j in range(len(y)):
      if x[i]**2 + y[j]**2 <= 1.0:
        inside[0].append(x[i])#add the x and y values to the empty list inside[0] and inside[1] respectively if the values are inside the radius
        inside[1].append(y[j])
      else:
        outside[0].append(x[i])#add the x and y values to the empty list outside[0] and outside[1] respectively if the values are outside the radius
        outside[1].append(y[j])
  
  pi = 4*len(inside[0]) / ((10**n)**2)#take the value of pi to the number of values inside the circle over the number of samples multiplied by 4

  plt.figure(figsize = (5,5))#create a figure to plot these x and y values - inside values are green and outised values are red
  plt.scatter(inside[0], inside[1], color = 'g')
  plt.scatter(outside[0], outside[1], color = 'r')
  plt.ylabel("$y$",fontsize=20)#making the font in latex form for the x and y axes labels
  plt.xlabel("$x$",fontsize=20)
  plt.title(f"{10**n} samples, pi = {pi}", fontsize = 30, color = 'r')#use fstring to print the values of pi and the number of samples for the title of each subplot
  plt.show()

  rerror = round((np.pi - pi)/np.pi,5)#using numpy.pi to determine the relative error 
  print(f"relative error = {rerror} \n")#use fstring to print the relative error of our pi underneath each plot and \n is used to indent it 
```

```
# Excersize 3
import numpy as np
from numpy import random as rd
import matplotlib.pyplot as plt

A = 1/(np.sqrt((2*np.pi)))#defining the constant of the formula 
def f(x):
  return A * np.exp((-x**2) / 2)#determing the function for the formula 
underf = 0#creating a counter for the integral
X = np.linspace(-1,1,1000)#set of values that will be used for the f(x) line 
for p in range(4):#using a for loop to obtain the power of the number of samples 
  samples = 10**(p)
  x = rd.uniform(low = -1, high = 1, size = 10**p)#numpy.random.uniform is used to obtain random values of x between -1 and 1 with 10**p samples
  
  y = rd.uniform(low = -1, high = 1, size = 10**p)#numpy.random.uniform is used to obtain random values of y between -1 and 1 with 10**p samples
  under, over = [[],[]] , [[],[]]
  for i in range(len(x)):#use two for loops to iterate all y the values over each f(x[i]) value for the y <= f(x) 
    for j in range(len(y)):
      if y[j] <= f(x[i]):
        underf += 1#the counter increases by one when y <=f(x)
        under[0].append(x[i])#add this x and y value to under[0] and under[1] respectively as the values are less than f(x)
        under[1].append(y[j])
      else:
        over[0].append(x[i])#otherwise these values are added to the lists that are above the f(x) line 
        over[1].append(y[j])
  
  integral = round(underf / (samples**2),4)#formula to determine the integral-which is the area under f(x) 

  plt.figure(figsize = (5,5))#plot a figue for each list of smaples and values under and above f(x) with the point above and purple and those below in orange
  plt.plot(X,f(X))#ploting f(X) to so we can see the actual function
  plt.scatter(under[0], under[1], color = 'orange')
  plt.scatter(over[0], over[1], color = 'purple')
  plt.ylabel("$y$",fontsize=20)#make the x and y axis labels large and in latex font
  plt.xlabel("$x$",fontsize=20)
  plt.title(f"{samples} samples, integral = {integral}", fontsize = 30)#make a large title for each plot giving the number of samples and the integral using fstring
  plt.show()
  rerror = round((0.68268 - integral)/0.68268,3)#place the relative error of the graph under the plot and indented to the left using an fstring nd the same formula as excersize 2 
  print(f"relative error = {rerror} \n")
```
