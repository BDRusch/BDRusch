```
#find the cross product of 3D vector
a = input("x coord: ")
b = input("y coord: ")
c = input("z coord: ")

d = input("2nd x coord: ")
e = input("2nd y coord: ")
f = input("2nd z coord: ")

r = []
r1 = eval(a)
r2 = eval(b)
r3 = eval(c)
r.extend((r1, r2, r3))
print(r)

p = []
p1 = eval(d)
p2 = eval(e)
p3 = eval(f)
p.extend((p1, p2, p3))
print(p)

def cross(r, p):
  y = [r2*p3 - r3*p2, r3*p1 - r1*p3, r1*p2 - r2*p1]
  return y

cross(r, p)


```

```
userin = input("what value to factorial: ")
def factorial(x):
  fact = 1
  if x < 0:
    print(f"please input a positive number")

  elif x >= 0:
    for x in range(2, x+1):
      fact = fact * x
    return fact

factorial(eval(userin))
```

```
Nuserin = input("what value for n: ")
Kuserin = input("what value for k: ")


def combine(n, k):
  if n < k:
    print(f"let n be greater than k")

  elif n >= k:
    result = factorial(n)/(factorial(k)*factorial(n - k))#using same function from above
  return result

combine(eval(Nuserin), eval(Kuserin))
```

```
#(x + y )^n
Xuserin = input("what value for x: ")
Yuserin = input("what value for y: ")
nuserin = input("what value for n: ")

def binomial(x, y, n):
  if n < 0:
    print("please input a positive n")# cant use binomial expansion on a negative n 

  elif n >= 0:
    terms = x**n# this is the first term of the binomial expansion
    for k in range(1, n + 1):
      terms = terms + (combine(n, k)*x**(n - k)*y**(k))
    return terms

binomial(eval(Xuserin), eval(Yuserin), eval(nuserin))
```
