```
import numpy as np 
from numpy import sin, cos
import matplotlib.pyplot as plt
def quadratic(a, b, c):
  Xval = np.linspace(-50, 50 , 1001)
  D = (b)**2 - 4*a*c
  rootD = D**(1/2)
  if D < 0:
    x0 = ((-1*b) + rootD)/(2*a)
    x1 = ((-1*b) - rootD)/(2*a)
    a = np.array([x0, x1])#create this array so that the complex numbers can be split into real and imaginary and then split up
    print(f"there are no real solutions but two complex solutions of x1 = {x0} and x2 = {x1}")
    plt.figure(figsize=(10, 10), dpi=100, facecolor='b')#dpi is resolution of figure
    t = np.linspace(0,2*np.pi,101)
    plt.plot(2*cos(t),2*sin(t))#makes a unit circle
    for x in range(len(a)):
      plt.plot([0,a[x].real],[0,a[x].imag],'ro-')#this is where the array "a" is split into imag and real parts and then plotted
    plt.xlabel('Real axis')
    plt.ylabel('Imaginary axis')
    plt.title('Quadratic Curve', fontsize= 20, color= 'purple')
    plt.xlim(-10, 10)#domain
    plt.ylim(-10, 10)#range
    plt.grid()#turns graph into grid
    plt.show()



  if D == 0:
    x2 = (-1*b)/(2*a)
    print(f"there is one solution of x = {x2}")
    Yval = [a*x**2 + b*x + c for x in Xval]
    plt.figure(figsize=(10, 10), dpi=100, facecolor='b')#dpi is resolution of figure
    plt.plot(Xval, Yval, 'purple', linewidth=2)#label each line so legend method can be used
    plt.plot([x2], [0], 'ro')
    plt.xlabel('xaxis')
    plt.ylabel('yaxis')
    plt.title('Quadratic Curve', fontsize= 20, color= 'purple')
    plt.xlim(-40, 40)#domain
    plt.ylim(-40, 100)#range
    plt.grid()#turns graph into grid
    plt.show()

  if D > 0:
    x3 = ((-1*b) + rootD)/(2*a)
    x4 = ((-1*b) - rootD)/(2*a) 
    print(f"there are two solutions x1 = {x3} and x2 = {x4}")
    Yval2 = [a*x**2 + b*x + c for x in Xval]
    plt.figure(figsize=(10, 10), dpi=100, facecolor='b')#dpi is resolution of figure
    plt.plot(Xval, Yval2, 'purple', linewidth=2)#label each line so legend method can be used
    plt.plot([x3], [0], 'ro')#makes a specific point a red dot
    plt.plot([x4], [0], 'ro')
    plt.xlabel('xaxis')
    plt.ylabel('yaxis')
    plt.title('Quadratic Curve', fontsize= 20, color= 'purple')
    plt.xlim(-40, 40)#domain
    plt.ylim(-40, 100)#range
    plt.grid()#turns graph into grid
    plt.show()

quadratic(1, 0, -1)
```

```
import numpy as np 
from numpy import sinh
import matplotlib.pyplot as plt

Temp = np.linspace(-5, 5, 100001)
Mag = np.isnan(((1 - (np.sinh(1/Temp))**4)**(1/8)))#isnan method shows area around zero
E = np.isnan(-1*(1 - (np.sinh(1/Temp))**4)**(-1*1/4))

plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)#subplot method (nrows, ncolumns, and then the graph youre working on)
plt.suptitle('Magnetization and Energy vs Temperature', fontsize= 20, color= 'b', y = 1.05)#supertitle for both subplots
plt.plot(Temp, Mag, 'b')
plt.xlabel('Temperature')
plt.ylabel('Mean Magnetization')
plt.xlim(-.5, 3)#domain
plt.ylim(-.5, 2)#range
plt.grid()#turns graph into grid

plt.subplot(1, 2, 2)
plt.plot(Temp, E, 'r')
plt.xlabel('Temperature')
plt.ylabel('Energy per Particle')
plt.xlim(-0.5, 3)#domain
plt.ylim(-0.5, 2)#range
plt.grid()#turns graph into grid
plt.tight_layout(pad = 3)#changes spacing between graphs so they dont overlap
plt.show()

```

```
import math
import numpy as np 
from numpy import e
import matplotlib.pyplot as plt
#fig, ax= plt.subplots(4, 2, figsize = (10,10))
fig, axes = plt.subplots(ncols=2,nrows=4, sharex=False, sharey=False, figsize = (10,10))
fig.suptitle('Taylor Series of e^x')
custom_xlim = (0, 20)
custom_ylim = (0, 3000)

# Setting the values for all axes.
plt.setp(axes, xlim=custom_xlim, ylim=custom_ylim)
def y(x, n):
  e_to_x = 0
  for i in range(n+1):
    e_to_x += x**i/math.factorial(i)
  return e_to_x
xval = np.linspace(0, 20, 1001)
yval = [y(xval, k) for k in range(8) ]

for i, ax in enumerate(axes.flatten()):
    ax.plot(xval, yval[i], 'r', label = r'f(x) = $e^{x}$')
    ax.plot(xval, np.e**xval, 'b', label= '$e^{x}$')
    ax.legend(loc="upper left")
    print(yval[i])
    
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

X, Y = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))

normp = np.sqrt((X+1)**2+Y**2)
normn = np.sqrt((X-1)**2+Y**2)

#Electric field of positive charge
Up = (X+1)/((normp)**3)
Vp = Y/((normp)**3)

#Electric field of negative charge
Un = -(X-1)/((normn)**3) #minus sign shows arrows point away from this point 
Vn = - Y/((normn)**3)

U = Up + Un
V = Vp + Vn

#Electric Field of Dipole
Un = U/np.sqrt(U**2+V**2)
Vn = V/np.sqrt(U**2+V**2)

plt.quiver(X, Y, Un, Vn, color='orange')
plt.title('Vector Field of Dipole x=-1 then Q=-1 x=1 then Q=1')
plt.show()
```

```
import numpy as np
import mpl_toolkits.mplot3d
from mpl_toolkits.mplot3d import Axes3D 
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (11, 11))
ax = fig.gca(projection = '3d')
X, Y, Z = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 10), np.linspace(-5,5, 10))

#charges above the origin dist from the centre
Top1_normp = np.sqrt((X+ 1)**2 + (Y+ 1)**2 + (Z+1)**2)#positive charges
Top2_normp = np.sqrt((X- 1)**2 + (Y- 1)**2 + (Z+1)**2)
Top1_normn = np.sqrt((X- 1)**2 + (Y+ 1)**2 + (Z+1)**2)#negative charges
Top2_normn = np.sqrt((X+ 1)**2 + (Y- 1)**2 + (Z+1)**2)

#charges below the origin dist from the centre 
Bottom1_normp = np.sqrt((X- 1)**2 + (Y+ 1)**2 + (Z- 1)**2)#positive charges
Bottom2_normp = np.sqrt((X+ 1)**2 + (Y- 1)**2 + (Z- 1)**2)
Bottom1_normn = np.sqrt((X+ 1)**2 + (Y+ 1)**2 + (Z- 1)**2)#negative charges
Bottom2_normn = np.sqrt((X- 1)**2 + (Y- 1)**2 + (Z- 1)**2)

#+ve charge(1, 1, 1)
U1 = (X+1)/((Top1_normp)**3)#U's are x direction
V1 = (Y+1)/((Top1_normp)**3)#V's are y direction
W1 = (Z+1)/((Top1_normp)**3)#W's are z direction

#+ve charge(-1, -1, 1)
U2 = (X- 1)/((Top2_normp)**3)
V2 = (Y- 1)/((Top2_normp)**3)
W2 = (Z+ 1)/((Top2_normp)**3)

#-ve charge(-1, 1, 1)
U3 = -(X- 1)/((Top1_normn)**3)
V3 = -(Y+ 1)/((Top1_normn)**3)
W3 = -(Z+ 1)/((Top1_normn)**3)

#-ve charge(1, -1, 1)
U4 = -(X+ 1)/((Top2_normn)**3)
V4 = -(Y- 1)/((Top2_normn)**3)
W4 = -(Z+ 1)/((Top2_normn)**3)

#+ve charge(-1, 1, -1)
U5 = (X- 1)/((Bottom1_normp)**3)
V5 = (Y+ 1)/((Bottom1_normp)**3)
W5 = (Z- 1)/((Bottom1_normp)**3)

#+ve charge(1, -1, -1)
U6 = (X+ 1)/((Bottom2_normp)**3)
V6 = (Y- 1)/((Bottom2_normp)**3)
W6 = (Z- 1)/((Bottom2_normp)**3)

#-ve charge(1, 1, -1)
U7 = -(X+ 1)/((Bottom1_normn)**3)
V7 = -(Y+ 1)/((Bottom1_normn)**3)
W7 = -(Z- 1)/((Bottom1_normn)**3)

#-ve charge(-1, -1, -1)
U8 = -(X- 1)/((Bottom2_normn)**3)
V8 = -(Y- 1)/((Bottom2_normn)**3)
W8 = -(Z- 1)/((Bottom2_normn)**3)


U = U1 + U2 + U3 + U4 + U5 + U6 + U7 + U8
V = V1 + V2 + V3 + V4 + V5 + V6 + V7 + V8
W = W1 + W2 + W3 + W4 + W5 + W6 + W7 + W8

#Electric Field of Dipole
Ex = U/np.sqrt(U**2+V**2+W**2)
Ey = V/np.sqrt(U**2+V**2+W**2)
Ez = W/np.sqrt(U**2+V**2+W**2)

ax.quiver(X, Y, Z, Ex, Ey, Ez, color='g', normalize = True)
plt.title('Vector Field of Octopole')
ax.view_init(50,30)
plt.show()
```
