```
# define function
def fizzbuzz (stop):
  """
  Fizzbuzz game that runs until argument stop. 
  """

  # instantiate variables (local)
  count = 1
  output = []

  # run while loop until count = stop
  while (count <= stop):
    # determine what to add to array based on rules
    if (count % 3 == 0 and count % 5 == 0):
      output.append("Fizzbuzz")
    if (count % 3 == 0):
      output.append("Fizz")
    if (count % 5 == 0):
      output.append("Buzz")
    else:
      output.append(str(count))

    count += 1 #increment counter

  # return array
  return output

# test function
fizzbuzz(15)
```

```
# define function
def calcSin (x, terms = 10):
  """
  Calculate the taylor expansion of sin(x) up to number of specified terms.
  """

  # define nested function
  def factorial(n):
    """
    Caluculate the factorial of n. 
    """
    # instantiate product (local)
    prod = 1

    # for loop to calculate factorial (runs from 1 to n)
    for i in range(1, n+1):
      prod *= i   #increment prod
    
    return prod
  # end nested function

  # initialise value for sinx (local)
  sinx = 0

  # loop to calculate each term (1, 3, 5, ...)
  for j in range(1, terms*2, 2):
    sinx += (-1)**(j//2) * x**j / factorial(j)

  return sinx

userx = eval(input("Which value of x in sin(x) do you want? "))  #asking the user for x and terms 
userterms = eval(input("How many terms should the expansion consist of? "))
usertaylor = calcSin(userx,userterms)  #activating the function

calcSin(userx, userterms)
```
