```python
# A Part

def LinearBetaMethod(Lambda , g , tRange , u0 , beta , h):
  sol = [u0]
  time = np.arange(tRange[0], tRange[-1], h)
  f = g(time)

  if beta== 0:
    for i in range(len(time)):
      un1 = (1 + h*Lambda)*sol[-1] + h*f[i + beta]
      sol.append(un1)

  if beta== 1:
    for l in range(len(time)-1):
      un1= (sol[-1] + h*f[l+1])/(1-h*Lambda)
      sol.append(un1)


  if beta == 0.5:
    for r in range(len(time)-1):
      mid=(f[r] + f[r+1])/2
      un1 = (2 + h*Lambda)*sol[-1]/(2-h*Lambda) + (2*h*mid)/(2-h*Lambda)
      sol.append(un1)
  
  return sol, time


Lambda = -1
g = lambda t: np.cos(t) - Lambda*np.sin(t)
h = 0.0001
u0 = 1
tRange=[0,2*np.pi]
sol1, time1 = LinearBetaMethod(Lambda , g , tRange , u0 , beta=1 , h=h)
sol2, time2 = LinearBetaMethod(Lambda , g , tRange , u0 , beta=0 , h=h)
sol3, time3 = LinearBetaMethod(Lambda , g , tRange , u0 , beta=0.5 , h=h)
correct = lambda t: np.exp(Lambda*t) + np.sin(t)

plt.figure()
plt.plot(time1, sol1, 'r', label = 'approximation (b=1)')
plt.plot(time2, sol2[0:-1], 'b', label = 'approximation (b=0)')
plt.plot(time3, sol3, 'k', label = 'approximation (b=0.5)')
plt.plot(time1, correct(time1), 'g', label = 'solution')
plt.legend()
plt.show()

print(len(time1))
print(np.round(2*np.pi/h))
```

```python
# B PART 

def newtonianfpnew(matrix, jacobian, x0, iteration, tol):
  X = []
  for g in range(len(x0)):
    if g==0:
      X = np.array([x0[g]])
    else:
      X = np.vstack((X, x0[g]))

  phi = lambda x: np.matmul(np.linalg.inv(jacobian(x)),matrix(x))
  ans = x0
  newval = []
  ERROR = []
  for y in range(len(x0)):
    newval.append(x0[y] - phi(x0)[y,0])
  XK1 = (np.vstack((x0, newval))).T

  for f in range(iteration-1):
    xuse = XK1[:, -1]
    fi = phi(xuse)
    valuse = []
    for t in range(len(xuse)):
      valuse.append(xuse[t]-fi[t, 0])
    valuse = np.transpose(np.array([valuse]))
    XK1 = np.hstack((XK1, valuse))
  
  for g in range(len(XK1[0, :])-1):
    vec = np.add(XK1[:, g],-XK1[:,g+1])
    vec = np.dot(vec, vec)
    ERROR.append(np.sqrt(vec))


  if ERROR[-1]<=tol:
    print('success')
    
  return XK1, ERROR

# Performs integration for the system of IVPs given by
# d/dt u = f(t, u), u(t_0) = u_0
# using the beta - method
# INPUT
# f the right - hand side function , the output should
# be a N x 1 array where N is the number of
# unknowns
# tRange the time interval [t_0 , T] of integration
# u0 initial solution at t = tRange (1) (N x 1 array )
# df a function that evaluates the Jacobian
# matrix of f
# beta defines the method
# h step - size
#
# OUTPUT
# tArray array containing the time points
# solArray array containing the solution at each time point
# (the ith row equals the solution at time
#  tArray (i))


def BetaMethod (f , tRange , u0 , df , beta , h ):
  tArray = np.arange(tRange[0], tRange[-1], h)
  un = []
  for e in range(len(u0)):
    if e==0:
      un = np.array([u0[e]])
    else:
      un = np.vstack((un, u0[e]))
  tol = 10**(-12)
  uvals = u0    #u0 is a row vector
  uvalz = []
  for i in range(len(tArray)):
    JacobianR = lambda x: np.identity(len(f(x, tArray[i])[:, 0])) - h*beta*df(x, tArray[i])
    if i==0:
      R = lambda x: -h*f(x, tArray[i])      
      un1, err = newtonianfpnew(R, JacobianR, u0, 50, tol)
      if err[-1]>tol:
        print('Error too large')
        break
      unvalz = un1[:,-1]
      un = np.hstack((un, unvalz))
    else:
      R = lambda x: un[:, -1] - un[:, -2] - h*f(x, tArray[i])
      un1, err = newtonianfpnew(R, JacobianR, un[:, -1], 50, 10**(-12))
      if err[-1]>tol:
        print('Error too large')
        break
      unvalz = un1[:, -1]
      un = np.hstack((un, unvalz))
  return un, tArray
      

t = [0, 10]
u00  = [1, 1]
f = lambda x, t: np.array([[Lambda*(x[0] - np.sin(t)) + np.cos(t)], [[Lambda*(x[1] - np.sin(t)) + np.cos(t)]]])
df = lambda x, t: np.array([[Lambda*x[0], Lambda], [Lambda*x[1], Lambda]])
 

BetaMethod(f, t, u00, df, 0, 1)
```
