```
#10
# y= ax^2 + bx + c

def cubic(y = 0, TF = True):

  a = input("input a")
  b = input("input b")
  c = input("input c")
  D = (eval(b))**2 - 4*eval(a)*eval(c)
  if TF == True:

    if D < 0:
      rootD = (D)**(1/2)
      x3 = ((-1*eval(b)) + (rootD))/(2*(eval(a)))
      x4 = ((-1*eval(b)) - (rootD))/(2*(eval(a)))
      print(f"The complex solutions are {x3} and {x4}")

    elif D > 0:
      rootD = (D)**(1/2)
      print("There are two solutions")
      x1 = ((-1*eval(b)) + (rootD))/(2*(eval(a)))
      x2 = ((-1*eval(b)) - (rootD))/(2*(eval(a)))
      print(f"x1 = {x1} and x2 = {x2}")

    elif D == 0:
      print("There is one solutions")
      x = (-1*eval(b))/(2*(eval(a)))
      print(x)

cubic()
```

```
#15
y = input("what number to factorialise:")
def factorial(x):
  fact = 1
  if x < 0:
    print("please input a positive number")
    
  elif x >= 0:
    for x in range(2, x+1):
      fact *= x
    return fact

  

factorial(eval(y))
```

```
#16
userin3 = input("give a value for n: ")
userin4 = input("give a value for k: ")
def combine(n, k):
  if n >= k :
    result = (factorial(n))/((factorial(k))*(factorial(n-k)))
    return result

  if n < k:
    print(" n must be greater than k")
combine(eval(userin3), eval(userin4))
  
```

```
#17
#(x+y)^n
m = input("value for x: ")
l = input("value for y: ")
o = input("value for n: ")
def binomial(x,y,n):
  if n >= 0:   #
    terms = x**n
    for k in range(1,n +1):
      terms += ((combine(n, k))*(x**(n-k))*(y**k))
      
    return terms

binomial(eval(m), eval(l), eval(o)) 
```
