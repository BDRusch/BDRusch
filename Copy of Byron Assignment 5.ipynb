```
import numpy as np 
from numpy import e
import matplotlib.pyplot as plt

def F(x):
  return (1 + (1/x))**x #function approximates e when x is very large

X = np.linspace(1,1000, 1000)
y = F(X)
k = np.e
tol=0.01
y_val=[]
for i in y:#made a for loop to find values in y where the difference between y and e is less than the tolerance
  if k - i < tol:
    y_val.append(i)
print(y_val[0])

x_val = np.interp(y_val[0], y, X)#this method finds the x value from X by using a corresponding y value such as y_val[0]
print(x_val)

x_ticks = np.arange(50, 550, 50)#gives the spacing of the xticks
print(x_ticks)

plt.figure(figsize=(12, 12))
plt.plot(X, y, 'blue', label= 'Approx for e at Large x')
plt.axhline(y = k, color='y', label='Actual Value e')#creates a horizontal line for y = e
plt.xlim(1,500)
plt.ylim(1,3)
plt.scatter(x_val, y_val[0], color = 'r', marker= 'o', label= 'Difference < Tolerance')#plots the point on the graph where the difference is less than tolerance
plt.legend()
plt.xticks(x_ticks)
plt.ylabel('y')
plt.xlabel('x')
plt.show()

```

```

import numpy as np
from numpy import e
import matplotlib.pyplot as plt

X = np.linspace(0,2,21)
print(X)
Y = ([np.e**i for i in X])
print(Y)

x = np.linspace(0,2,20)
y = derivative(X, Y)

plt.figure(figsize=(12, 12))
plt.plot(X, Y, 'r', label='x vs $e^x$')
plt.plot(x, y, 'purple', label='x vs $dy/dx$  $e^x$')
plt.legend()
plt.show()

def symmetric_derivative(x, f):
  #this will be a function to find the derivative
  diff = np.ones(len(x) - 1)#makes an array full of zeroes which can be changed with a simple formula or something else
  dx = [(x[i + 1] - x[i]) for i in range(len(x) - 1)]#makes a list of delta x - use len(x)-1 as there will be one less difference than there is data points
  for i in range(1, len(x) - 1):
    diff[i] = (f[i + 1] - f[i - 1])/(2*dx[i])#formula for the derivative of a function - this time range starts from 1 due to f[i - 1]
                                              #everything else stays the same but changed diff = np.ones instead of np.zeroes
  return diff#composes into a list

x = np.linspace(0,2,20)
y = derivative(X, Y)
print(symmetric_derivative(X, Y))
y1 = symmetric_derivative(X, Y)

plt.figure(figsize=(12, 12))
plt.plot(x, y1, 'blue', label='x vs $dy/dx$ $e^x$ (Symmetric Diff)')
plt.plot(x, y, 'orange', label='x vs $dy/dx$  $e^x$ (Newtons Diff)')
plt.plot(X, Y, 'g', label='x vs $e^x$')
plt.legend()
plt.show()

```

```
import numpy as np
from numpy import cos
import matplotlib.pyplot as plt

x = np.linspace(0,10,101)
y = ([np.cos(i) for i in x])
print(y)

plt.figure(figsize=(10,10))
plt.plot(x, y, 'g', label='cosx')

def nth_derivative(x, f, n):
  #this will be a function to find the nth derivative
  p = 1
  diff = f
  while p <= n :
    dx = [(x[i + 1] - x[i]) for i in range(len(x)- p)]#makes a list of delta x - use len(x)-p 
    for i in range(len(x) - p):                       #as there will be one less difference than there is data points for each consecutive derivation
      diff[i] = (diff[i + 1] - diff[i])/dx[i]#formula for the derivative of a function inside the while loop so the diff list is changed each loop
    p = p + 1#change the value of p to restart the derivation if p<=n
  print(diff)
  return diff#composes into a list
    
Y = nth_derivative(x, y, 4)
Y2 = nth_derivative(x[:-4], Y[:-4], 4)
plt.plot(x[:-4],Y[:-4], 'purple', label='cosx Derived 4 Times')
plt.plot(x[:-8],Y2[:-4], 'r', label='cosx Derived 8 Times')
plt.legend()
plt.show()
```

```
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt


f = lambda x: x**2 - 2 #function to integrate
a = -2#bounds of integration
b = 2

intExact, err = integrate.quad(f, a, b)
print(intExact)#prints the actaul area under the graph 

x = np.linspace(a, b, 100)#x values for the line
y = ([(i**2 - 2) for i in x])#y values for the line using list comprehension

fig, ax = plt.subplots(3, 3, figsize = (9, 9), constrained_layout= True)#making 9 subplots in 3 rows and columns
plt.suptitle('Midpoint Rule Integration')


    

dx=[]
x1=[]
x2=[]

for N in range(1, 10):#for loop to create separation of endpoints 
  del_x = (b - a)/N#gives separation of end points 
  dx.append(del_x)#add to an empty list each time a new separation is found
  X1 = np.linspace(a, b, N + 1)#find the number of point - use N+1 as for the the first one there are two point -2 and 2
  x1.append(X1)

print(dx)
print(x1)



x2_mid = [[(i[j+1]+i[j])/2 for j in range(len(i)-1)] for i in x1] #nested for loop in list comprehension to create the midpoints of x
print(x2_mid)

y2_mid = [[(i[j]**2 - 2) for j in range(len(i))] for i in x2_mid] #calculates the height at those midpoints
print(y2_mid)

Area = []
for i, item in enumerate(y2_mid):
  Area.append(list(map(lambda x: x*dx[i], item)))#use i, item in enumerate(y2_mid) to create a list of areas of the bars 
print(Area)                                       #dx[i]uses the correct dx corresponding to the yvalues which then multiplied by each x in item as y2_mid is a list of lists 
                                                  # thus creating a new list of lists and Area=[] appends each small list to its empty list
Area_final1 = [sum(x) for x in Area]# list comprehension used to sum up the individual lists inside the Area list 
print(Area_final1)                  #creates a list of 9 values which are the areas under the bars for each graph

Area_positive = []                          #same method used to create the area list but no want to turn each element inside every list to +ve value
for i, item in enumerate(Area):             #use lambda function which gives absolute value of each value in 'item'
  Area_positive.append(list(map(lambda x: abs(x), item)))#as again Area is a list of lists thus 'item' is a list//thus after this process = list of list
print(Area_positive)

Area_final2 = [sum(x) for x in Area_positive]
print(Area_final2)

Area_dp = ['%.2f'% i for i in Area_final1]# method of making all elements in Area_final1 to 2 decimal places using'%.2f'%
print(Area_dp)

ax = ax.flatten()#flatten to make it easier to edit the subplots
for i in range(9):
  ax[i].plot(x, y, color='black', label = f'f(x) = $x^2$ - 2 for N = {i+1}')
  ax[i].bar(x2_mid[i], y2_mid[i], dx[i], alpha = 0.5, color= 'g', edgecolor = 'r')
  ax[i].title.set_text(f'Computed Surface Area = {Area_dp[i]}')
  ax[i].legend()



plt.show()
    
```
