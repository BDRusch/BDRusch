```
# Refractive index against time


import matplotlib.pyplot as plt
import numpy as np


 # Found with symfit.

x = np.array([79,92,101,102,103,130])
y = np.array([1.00, 1.21, 1.13, 1.23,1.16, 1.31])

# order : air, white wine, oil, vinegar , sugar water , water


a, c = np.polyfit(x, y, 1, cov = True)

print("The error of the slope is {}".format(np.sqrt(c[0][0])))  # first term of the covariance matrix
print("The error of the y-intercept is {}".format(np.sqrt(c[1][1])))    #last term of the covariance matrix

m, b = np.polyfit(x, y, 1)


gradient = m

y_intercept =b

# From here the plotting starts

xerr = [30,30,30,0.1,30, 30]     # enter errors of time
yerr = [0.04,0.04,0.04,0.05,0.04,0.04]      # errors of refractive index


fig, ax = plt.subplots()


ax.errorbar(x, y,
            xerr=xerr,
            yerr=yerr,
            fmt='o', markersize='10')


plt.xlabel('Time Taken (ns)')
plt.ylabel('Refractive Index')
plt.title('Refractive Index against Time Taken')
plt.plot(x, m*x + b, label='0.005235252758605843*x+0.6428590874957555')
plt.legend()
plt.show()
```

```
# Density vs time taken

import matplotlib.pyplot as plt
import numpy as np


 # Found with symfit.

x = np.array([79,92,101,102,103,130])     
y = np.array([1.225, 1000.000, 870.000, 900.000,1400.000, 1000.000])    

# order : air, white wine, oil, vinegar , sugar water , water

a, c = np.polyfit(x, y, 1, cov = True)

print("The error of the slope is {}".format(np.sqrt(c[0][0])))  # first term of the covariance matrix
print("The error of the y-intercept is {}".format(np.sqrt(c[1][1])))    #last term of the covariance matrix

m, b= np.polyfit(x, y, 1)

gradient = m

y_intercept =b

# From here the plotting starts

xerr = [30,30,30,0.1,30, 30]# enter errors for time
yerr = [0,200,70,200,300,100]# errors for density


fig, ax = plt.subplots()


ax.errorbar(x, y,
            xerr=xerr,
            yerr=yerr,
            fmt='o')


print(gradient)
print(y_intercept)

plt.xlabel('Time Taken (ns)')
plt.ylabel('Density (kg/m^3)')
plt.title('Density vs Time Taken')
plt.plot(x, m*(x) + b, label='15.453912628603401*x -706.3651246813303' )
plt.legend()
plt.show()



```
