```
#20
integers = [1]
for integer in integers :
  integer += 1
  integers.append(integer)
  if integer == 100:
    break

print(integers)

squares = [1]
for square in squares :
  square = ((((square)**(1/2))+ 1)**2)
  squares.append(square)
  if square >= 100 :
    break

print(squares)

intnotinsqrs = []
for x in integers:
  if x not in squares:
    intnotinsqrs.append(x)

print(intnotinsqrs)
```
