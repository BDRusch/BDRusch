```
import numpy as np

vector_a = [1,2,3]
square = 0

#squaring each item in vector and adding to 'square'
for i in range(3):
  square += vector_a[i] * vector_a[i]
print('vector_a squared = ', square)

#vector_a dot vector_b
vector_b = [0,-5,6]
product = 0
for i in range(3):
  product += vector_a[i] * vector_b[i]
print('vector_a dot vector_b = ', product)

#vector_c dot vector_d
vector_c = list(range(10))
vector_d = list(range(10,20))[::-1]
product = 0
for i in range(10):
  product += vector_c[i] * vector_d[i]
print('vector-c dot vector_d = ', product)

###vector_a dot vector_d
#over index a
product = 0
for i in range(3):
  product += vector_a[i] * vector_d[i]
print('v_a dot v_d over index_a = ', product)

#over index d
product = 0
for i in range(10):
  if i == len(vector_a):   #stops code before i exceeds index_a
    break
  product += vector_a[i] * vector_d[i]
print('v_a dot v_d over index_d = ', product)

#dot product over index_a gives mathematically correct answer, as the excess index_d values multiply with 0
#multiplying over index_d gives an error, though if stopped just before would give right answer
```

```
#defining cumsum
def cumsum(list):     #creating function for cumsum based on arbitrary list
  sumcum = [list[0]]    #sumcum is the list of cumulatative sums, starting with the first element in the list
  for i in range(len(list)):  #for loop thatll iterate as many times as there are items in the list
    if i == (len(list)-1):    #stopping code before list[i+1] exceeds len(list)
      break
    sumcum.append(sumcum[i] + list[i+1])  #(sumcum[i] + list[i+1]) makes the newest cumulatative sum and a
  return sumcum


#defining integers
integers = [i+1 for i in range(100)]   #i+1 to have integers start from 1 rather than 0 (could also do range(1,100,1))
print("integers: ", integers) 
print("cumsum: ", cumsum(integers)) #printing cumsum
print()

#defining alt_integers
alt_integers = [i for i in range(1,1000,2)] #odd integers starting from 1
print("alt_integers:", alt_integers)
print("cumsum: ", cumsum(alt_integers)) #printing cumsum
print()

#defining fibonacci
fibonacci = [1,1]
for i in range(100):
  fibonacci.append(fibonacci[i]+fibonacci[i+1])
print("fibonacci: ", fibonacci)
print("cumsum: ", cumsum(fibonacci)) #printing cumsm
```
