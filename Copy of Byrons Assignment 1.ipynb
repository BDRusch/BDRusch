```
integers = [ 1 ]
for i in integers : # for loop that prints numbers in a list from 1 - 100
  i = i + 1
  integers.append(i)
  if i == 100 :
    break #break tells for loop to stop looping
print(integers)

squares = [ 1 ]
for squ in squares:
  squ = (((squ)**0.5) + 1)**2 # formula to count next square number
  squares.append(squ)
  if squ == 100 :
    break
print(squares)

intnotinsqurs = []
for x in integers :
  if x not in squares :
    intnotinsqurs.append(x)

print(intnotinsqurs)


```

```
#using list comprehension for same work as above
intlist = [ x for x in range(1,100)]
print(intlist)
squareslist = [ x**2 for x in range(1,11)]
print(squareslist)
newlist = [ x for x in intlist if x not in squareslist]
print(newlist)
```

```
userin = input("what is a large number")
fizzbuzz = []
for x in range(1,eval(userin)):
  if x % 3 == 0 and x % 5 != 0 :
    x = "Fizz"
    fizzbuzz.append("Fizz")

  elif x % 5 == 0 and x % 3 != 0 :
    x = "Buzz"
    fizzbuzz.append("Buzz")

  elif x % 3 == 0 and x % 5 == 0:
    x = "FizzBuzz"
    fizzbuzz.append("FizzBuzz")
  else: fizzbuzz.append(x)

print(fizzbuzz)
```
